```python
#ENRIQUECER O DATASET tb_plt COM INFORMAÇÕES BASEADAS NOS DADOS COLETADOS.

#################### Marcar os clientes que mês que tiveram Churn

#Identifica a planta atual, e a ultima referencia (formada por ANO e MES)
ultimo_mes = tb_plt['REFERENCIA'].max()

# Converte a coluna DATA_INSTALACAO em data
tb_plt['DATA_INSTALACAO'] = pd.to_datetime(tb_plt['DATA_INSTALACAO'])

#Tratando o COD_ID para relacão entre meses
tb_plt['COD_CHAVE_CLIENTE'] = 'C-'+tb_plt['COD_ID'].str[5:-4]

# Gera o máximo de REFERENCIA por COD_CHAVE_CLIENTE
max_referencia = tb_plt.groupby('COD_CHAVE_CLIENTE')['REFERENCIA'].max().reset_index()

# Renomea a coluna RESULTADO para MAX_REFERENCIA
max_referencia = max_referencia.rename(columns={'REFERENCIA': 'MAX_REFERENCIA'})

# Junta o dataframe original com a coluna MAX_REFERENCIA
tb_plt = pd.merge(tb_plt, max_referencia, on='COD_CHAVE_CLIENTE')

# Cria a coluna CHURN com o valor 0 defaul
tb_plt['CHURN'] = 0

# Marca os clientes no mês que tiveram CHURN - Sairam da planta
tb_plt.loc[tb_plt['REFERENCIA'] == tb_plt['MAX_REFERENCIA'], 'CHURN'] = 1

# Desmarca os clientes da última referência, pois contiuam na planta.
tb_plt.loc[tb_plt['REFERENCIA'] == ultimo_mes, 'CHURN'] = 0

# Função para calcular a diferença em meses entre dois valores de ano e mês
def diferenca_meses(referencia_str, data_instalacao_dt):
    referencia_dt = datetime.strptime(referencia_str, '%Y%m')
    return (referencia_dt.year - data_instalacao_dt.year) * 12 + (referencia_dt.month - data_instalacao_dt.month)

# Converter a coluna REFERENCIA para o formato YYYYMM e criar a coluna DATA_INSTALACAO_DT com a data de instalação em formato datetime.date
tb_plt['DATA_INSTALACAO_DT'] = tb_plt['DATA_INSTALACAO'].dt.date

# Criar a coluna MESES_COMO_CLIENTE com a diferença em meses entre REFERENCIA e DATA_INSTALACAO_DT
tb_plt['MESES_COMO_CLIENTE'] = [diferenca_meses(ref_str, inst_dt) for ref_str, inst_dt in zip(tb_plt['REFERENCIA'], tb_plt['DATA_INSTALACAO_DT'])]

# Criar a coluna EXPERIENCIA_CLIENTE com base na coluna MESES_COMO_CLIENTE, marcando como NOVO caso seja <=3 meses, se não, MADURO.
tb_plt['EXPERIENCIA_CLIENTE'] = tb_plt['MESES_COMO_CLIENTE'].apply(lambda x: 'NOVO' if x < 4 else 'MADURO')

#############################
```

```python
# Marca a coluna TIPO_IMOVEL como PREDIO OU CASA
tb_plt['TIPO_IMOVEL'] = np.where(tb_plt['COMPLEMENTO'].fillna('').str.contains('AP', case=False), 'PREDIO', 'CASA')

# Marca a coluna TOTAL_PLANTA = 1
tb_plt['TOTAL_PLANTA'] = 1

# Exclui colunas que não serão utilizadas.
tb_plt = tb_plt.drop(columns=['CLUSTER','MAX_REFERENCIA', 'COMPLEMENTO'], axis=1)
```

```python
#ENRIQUECER O DATASET tb_reparos COM INFORMAÇÕES BASEADAS NOS DADOS COLETADOS.

#Identifica o primeiro mes, a primeira referencia (formada por ANO e MES).
primeiro_mes = tb_reparo['REFERENCIA'].min()

# Converte a coluna DATA_CRIACAO em data
tb_reparo['DATA_CRIACAO'] = pd.to_datetime(tb_reparo['DATA_CRIACAO'])

#Agrupando ID_CLIENTE E DATA_INSTALACAO
tb_plt_agrupado = tb_plt.groupby(['COD_CHAVE_CLIENTE', 'DATA_INSTALACAO']).size().reset_index().iloc[:, :2]

#Realizando o LEFT JOIN entre tb_reparos, tb_plt_agrupado
tb_reparo = pd.merge(tb_reparo, tb_plt_agrupado[['COD_CHAVE_CLIENTE', 'DATA_INSTALACAO']], on='COD_CHAVE_CLIENTE', how='left')

#Ordenando pela DATA_CRIACAO
tb_reparo = tb_reparo.sort_values(by='DATA_CRIACAO')

tb_reparo['REPETIDA_30_DIAS'] = tb_reparo.groupby('COD_CHAVE_CLIENTE')['DATA_CRIACAO'].apply(lambda x: (x - x.shift(1)).dt.days <= 30).astype(int)

#Marcando 1 para TOTAL DE REPAROS
tb_reparo['TOTAL_REPAROS'] = 1

# Desmarcar os clientes da primeira referência, pois não tem uma safra completa de 30 dias
tb_reparo.loc[tb_reparo['REFERENCIA'] == primeiro_mes, 'REPETIDA_30_DIAS'] = 0
```

```python
# Excluindo atributos auxiliares já utilizados ou que não serão utilizadas nas proximas etapas.
# Transformando os campos TOTAL_REPAROS E REPETIDA_30_DIA em tipo inteiro.
tb_plt = tb_plt.drop(columns=['COD_CHAVE_CLIENTE','DATA_INSTALACAO', 'COD_CHAVE_CLIENTE','MESES_COMO_CLIENTE','DATA_INSTALACAO_DT'], axis=1)
tb_reparo = tb_reparo.drop(columns=['COD_CHAVE_CLIENTE','COD_REPARO','DATA_CRIACAO','DATA_INSTALACAO'], axis=1)
tb_reparo['TOTAL_REPAROS'] = tb_reparo['TOTAL_REPAROS'].fillna(0).astype(int)
tb_reparo['REPETIDA_30_DIAS'] = tb_reparo['REPETIDA_30_DIAS'].fillna(0).astype(int)
```

```python
# Agrupando o COD_ID e sumarizando o Total de reparos e Repetidas 30 dias.
tb_reparos_sumarizado_cod_id = tb_reparo.groupby('COD_ID').agg({'TOTAL_REPAROS': 'sum', 'REPETIDA_30_DIAS': 'sum'})

#Realizando o LEFT JOIN entre tb_plt x tb_reparos_sumarizado_cod_id para marcar o volume de reparos/repetidos.
tb_plt_reparo = pd.merge(tb_plt, tb_reparos_sumarizado_cod_id, on='COD_ID', how='left')
tb_plt_reparo['TOTAL_REPAROS'] = tb_plt_reparo['TOTAL_REPAROS'].fillna(0).astype(int)
tb_plt_reparo['REPETIDA_30_DIAS'] = tb_plt_reparo['REPETIDA_30_DIAS'].fillna(0).astype(int)
```

```python
# Importe os módulos necessários
from pyspark.sql import SparkSession

# Crie uma SparkSession
spark = SparkSession.builder.appName("pyspark-notebook").master("local[*]").getOrCreate()
```

```python
#criando os DataFrames em Spark 
plt_reparo_spark = spark.read.csv(path="/content/tb_plt_reparo_trat.csv", header=True)
rede_spark = spark.read.csv(path="/content/drive/MyDrive/PROJCD/tb_rede.csv", header=True)
dispositivo_spark = spark.read.csv(path="/content/drive/MyDrive/PROJCD/tb_dispositivo.csv", header=True)
plt_reparo_spark = plt_reparo_spark.drop("_c0")
rede_spark = rede_spark.drop("_c0")
dispositivo_spark = dispositivo_spark.drop("_c0")
```

```python
#criando as visões temporárias 
plt_reparo_spark.createOrReplaceTempView("tb_plt_reparo_spark")
rede_spark.createOrReplaceTempView("tb_rede_spark")
dispositivo_spark.createOrReplaceTempView("tb_dispositivo_spark")
```

```python
# QUERY KPIs CLUSTER PILOTO

kpi= """
SELECT 'CLUSTER_PILOTO' REGIAO,
       ROUND(SUM(TOTAL_PLANTA)/12) AS SHARE_PLANTA,
       ROUND(SUM(CHURN) / SUM(TOTAL_PLANTA) * 100, 2) AS PERC_CHURN, 
       ROUND(SUM(TOTAL_REPAROS) / SUM(TOTAL_PLANTA) * 100, 2) AS PERC_IRT, 
       ROUND(SUM(REPETIDA_30_DIAS) / SUM(TOTAL_REPAROS) * 100, 2) AS PERC_IRR

FROM tb_plt_reparo_spark
WHERE REFERENCIA >= 202201 AND REFERENCIA < 202301
"""
spark.sql(kpi).show()
df_spark = spark.sql(kpi)
kpi = df_spark.toPandas()

df= kpi
```

```python
# GRAFICO - KPIs CLUSTER PILOTO

# criação do gráfico de barras 
# define as dimensões da figura e chama a função 'barplot' para criar o gráfico, passando as informações de x, y e hue.
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(x='REGIAO', y='value', hue='variable', 
            data=pd.melt(df, id_vars='REGIAO',  value_vars=['PERC_CHURN', 'PERC_IRT', 'PERC_IRR']), ax=ax)

# Adição dos valores acima das barras e ajuste da legenda
# função 'annotate' para adicionar os valores de 'PERC_CHURN', 'PERC_IRT' e 'PERC_IRR' nas posições corretas. 
# também é feita a formatação dos valores para apresentar apenas duas casas decimais e adicionar o símbolo de porcentagem.
for i, row in df.iterrows():
    ax.annotate("{:.2f}%".format(row['PERC_CHURN']), xy=(i-0.2, row['PERC_CHURN']+1), ha='center', va='bottom')
    ax.annotate("{:.2f}%".format(row['PERC_IRT']), xy=(i, row['PERC_IRT']+1), ha='center', va='bottom')
    ax.annotate("{:.2f}%".format(row['PERC_IRR']), xy=(i+0.2, row['PERC_IRR']+1), ha='center', va='bottom')
    ax.set_title('KPI´s - CLUSTER PILOTO', fontweight='bold', fontsize=14)
    ax.set_ylim(0, 15)  
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.set_yticks([])
    ax.legend(title='',  loc='upper left')
# exibição do gráfico
plt.show()
```

```python
# QUERY KPIs POR CIDADE

kpi_cidade= """
SELECT CIDADE,
       ROUND(SUM(TOTAL_PLANTA) / SUM(SUM(TOTAL_PLANTA)) OVER () * 100, 2) AS SHARE_PLANTA,
       ROUND(SUM(CHURN) / SUM(TOTAL_PLANTA) * 100, 2) AS PERC_CHURN, 
       ROUND(SUM(TOTAL_REPAROS) / SUM(TOTAL_PLANTA) * 100, 2) AS PERC_IRT, 
       ROUND(SUM(REPETIDA_30_DIAS) / SUM(TOTAL_REPAROS) * 100, 2) AS PERC_IRR

FROM tb_plt_reparo_spark
WHERE REFERENCIA >= 202201 AND REFERENCIA < 202301
GROUP BY CIDADE 
ORDER BY 2 DESC
"""
spark.sql(kpi_cidade).show()
df_spark = spark.sql(kpi_cidade)
kpi_cidade = df_spark.toPandas()

df= kpi_cidade
```

```python
# GRAFICO - KPIs POR CIDADE

ig, (ax4, ax) = plt.subplots(1, 2, figsize=(20, 5))

# Cria o gráfico de pizza (pie chart) com a participação percentual de cada cidade em relação ao total de plantas.
# Método countplot() da biblioteca Seaborn para gerar o gráfico de pizza
ax4.clear()
ax4.set_title('%SHARE PLANTA', fontweight='bold', fontsize=14)
ax4.set_xlabel('CIDADE')
ax4.pie(df['SHARE_PLANTA'], labels=df['CIDADE'], autopct='%1.1f%%', wedgeprops=dict(width=0.6), radius=1, center=(0, 0))

# Cria o gráfico de barras mostrando os valores percentuais de churn, IRT e IRR para cada cidade. 
# Método barplot() para o gráfico de barras. 
sns.barplot(x='CIDADE', y='value', hue='variable', data=pd.melt(df, id_vars='CIDADE', 
                                                                value_vars=['PERC_CHURN', 'PERC_IRT', 'PERC_IRR']), ax=ax)
funções 
for i, row in df.iterrows():
    ax.annotate("{:.2f}%".format(row['PERC_CHURN']), xy=(i-0.2, row['PERC_CHURN']+1), ha='center', va='bottom')
    ax.annotate("{:.2f}%".format(row['PERC_IRT']), xy=(i, row['PERC_IRT']+1), ha='center', va='bottom')
    ax.annotate("{:.2f}%".format(row['PERC_IRR']), xy=(i+0.2, row['PERC_IRR']+1), ha='center', va='bottom')
    ax.legend(title='')

ax.set_title('CHURN, IRT e IRR por CIDADE', fontweight='bold', fontsize=14)
ax.set_xlabel('CIDADE')
ax.set_ylabel('')
ax.set_ylim(0, 15)  
ax.xaxis.set_tick_params(rotation=45) 
ax.set_yticks([])

# Calcula os KPIs do Cluster Piloto
tb_plt_reparo['REFERENCIA'] = tb_plt_reparo['REFERENCIA'].astype(int)
mean_churn = round((tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['CHURN'].sum() 
/ tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['TOTAL_PLANTA'].sum()) * 100, 1)
mean_irt = round((tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['TOTAL_REPAROS'].sum() 
/ tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['TOTAL_PLANTA'].sum()) * 100, 1)
mean_irr = round((tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['REPETIDA_30_DIAS'].sum() 
/ tb_plt_reparo.query('REFERENCIA >= 202201 and REFERENCIA < 202301')['TOTAL_REPAROS'].sum()) * 100, 1)

# Desenha as linhas
ax.axhline(mean_churn, color='blue', linestyle='-.', linewidth=1)
ax.axhline(mean_irt, color='#FF8C00', linestyle='-.', linewidth=1)
ax.axhline(mean_irr, color='green', linestyle='-.', linewidth=1)

plt.show()
```
